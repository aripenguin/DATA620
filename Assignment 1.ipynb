```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
G = nx.Graph()

G.add_nodes_from(['Andre', 'Carol', 'Fernando', 'Diane', 'Beverly', 'Garth', 'Ed', 'Heather', 'Ike', 'Jane'])

G.add_edges_from([
    ('Andre', 'Carol'), ('Andre', 'Diane'), ('Andre', 'Beverly'), ('Andre', 'Fernando'),
    ('Carol', 'Fernando'), ('Carol', 'Diane'), ('Diane', 'Fernando'), ('Diane', 'Garth'), ('Diane', 'Beverly'), ('Diane', 'Ed'),
    ('Beverly', 'Garth'), ('Beverly', 'Ed'), ('Ed', 'Garth'),
    ('Garth', 'Fernando'), ('Garth', 'Heather'), ('Heather', 'Fernando'), ('Heather', 'Ike'), ('Ike', 'Jane')])

nx.draw(G, with_labels=True, node_color='skyblue', edge_color='gray')
plt.show() 
```
